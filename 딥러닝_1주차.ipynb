1.4.2 클래스

```python
# Man 클래스 선언
class Man:
    # 클래스 내부의 함수는 첫 번째 매개변수로 반드시 self 입력, self는 자기 자신을 나타내는 dictionary임.
    # __init__ : constructor(생성자) method, 반드시 첫 번째로 써야 함. self.name=name은 전달받은 내용을 객체 내부에 저장하는 역할.
    def __init__ (self,name,age):
        self.name=name
        self.age=age
        print(f"{self.name} is born")

    # speak : 클래스 내부의 일반 매서드.
    def speak(self):
        print(f"{self.name} is speaking")

    def aging(self):
        print(f"{self.age} years old!")

# instance 이름 = 클래스("aaaa") : 붕어빵 틀인 클래스에 aaaa를 넣어서 실체화된 붕어빵을 만든다.
m=Man("David",27)
m.speak()
m.aging()
```

1.6.1 matplotlib 단순한 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,6,0.1)
y=np.sin(x)

plt.plot(x,y)
plt.show()
```

1.6.2 pyplot의 기능

```python
x=np.arange(0,6,0.1)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,label='sin')
plt.plot(x,y2,linestyle='--',label='cos')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('sin and cos Graph')
plt.legend()
plt.show()
```
